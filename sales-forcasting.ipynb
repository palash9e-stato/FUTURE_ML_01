```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Order Date'].head()
```

```python
df[df['Sales'] > 0].head()
```

```python
monthly_sales = (df.groupby(pd.Grouper(key='Order Date', freq='ME'))['Sales'].sum().reset_index())
```

```python
monthly_sales.rename(columns={'Order Date': 'ds','Sales': 'y'}, inplace=True)
```

```python
monthly_sales.head()
```

```python
Q1 = monthly_sales['y'].quantile(0.25)
Q3 = monthly_sales['y'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
monthly_sales['y'] = monthly_sales['y'].clip(lower=lower_bound, upper=upper_bound)
monthly_sales.head()
```

```python
from prophet import Prophet

model = Prophet(yearly_seasonality=True, weekly_seasonality=False, daily_seasonality=False, changepoint_prior_scale=0.1)
model.add_seasonality(name='quarterly', period=91.25, fourier_order=8)
model.fit(monthly_sales)
```

```python
future_predictions = model.make_future_dataframe(periods=6, freq='ME')
Predicted_Sales = model.predict(future_predictions)
Predicted_Sales.head()
```

```python
forecast_output = Predicted_Sales[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].merge(monthly_sales, on='ds', how='left')
```

```python
forecast_output.rename(columns={'ds': 'Date', 'y': 'Actual_Sales', 'yhat': 'Forecast_Sales', 
                                'yhat_lower': 'Lower_Bound', 'yhat_upper': 'Upper_Bound'}, inplace=True)
forecast_output.head()
```

```python
forecast_output['Error'] = (
    forecast_output['Actual_Sales'] - forecast_output['Forecast_Sales']
)

forecast_output['APE_%'] = (
    forecast_output['Error'] / forecast_output['Actual_Sales']
) * 100
```

```python
Predicted = forecast_output.Forecast_Sales
Actual = forecast_output.Actual_Sales
```

```python
Actual.isna().sum()
```

```python
valid_data = forecast_output.dropna(subset=['Actual_Sales'])
```

```python
Actual = valid_data['Actual_Sales']
Predicted = valid_data['Forecast_Sales']
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
```

```python
r2score = r2_score(Actual, Predicted)
r2score
```

```python
import math
mean_squared_error(Actual, Predicted)
```
